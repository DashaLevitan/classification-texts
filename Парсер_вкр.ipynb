```python
from bs4 import BeautifulSoup
import requests
from concurrent.futures import ThreadPoolExecutor
from docx import Document
import pdfminer.high_level
from pdfminer.high_level import extract_text
import pandas as pd
import re

# Создаем пустой DataFrame
df = pd.DataFrame(columns=['topic', 'score', 'faculty', 'text'])

#url = 'https://www.hse.ru/edu/vkr/?language=ru&text_available=yes'

# Прикладная математика
#url = 'https://www.hse.ru/edu/vkr/?program=p135181813%3Bg122468467%3Bg134854846%3Bp135181638&language=ru&text_available=yes'

# Прикладная математика и информатика
#url = 'https://www.hse.ru/edu/vkr/?program=g122722324%3Bg134854782%3Bp135181588%3Bp220156607%3Bp135181634%3Bg122392182%3Bp225211210%3Bp135181636&language=ru&text_available=yes'

# Анализ данных в биологии и медицине
#url = 'https://www.hse.ru/edu/vkr/?program=p164325560&language=ru&text_available=yes'

# Бизнес-информатика
#url = 'https://www.hse.ru/edu/vkr/?program=g122392179%3Bp135181726%3Bp135181468%3Bp135181722%3Bp377879003%3Bg122468449%3Bp581739127%3Bp377879230%3Bp135181724%3Bg122468450%3Bg52587428%3Bg52587427%3Bp135181460&language=ru&text_available=yes'

# Бизнес-информатика (в)
#url = 'https://www.hse.ru/edu/vkr/?program=p135181738&language=ru&text_available=yes'

# Инжиниринг в электронике
#url = 'https://www.hse.ru/edu/vkr/?program=g74148977%3Bp135181374&language=ru&text_available=yes'

# ИАД
#url = 'https://www.hse.ru/edu/vkr/?program=p141401855&language=ru&text_available=yes'

# ИТСС
#url = 'https://www.hse.ru/edu/vkr/?program=p135181652%3Bg122468462&language=ru&text_available=yes'

# ИВТ
#url = 'https://www.hse.ru/edu/vkr/?program=p135181640%3Bg122468466%3Bg53189678&language=ru&text_available=yes'

# Информационная аналитика
#url = 'https://www.hse.ru/edu/vkr/?program=p581324076%3Bp135181470&language=ru&text_available=yes'

# ИБ
#url = 'https://www.hse.ru/edu/vkr/?program=p225178429&language=ru&text_available=yes'

# Информационные системы и взаимодействие человек-компьютер
#url = 'https://www.hse.ru/edu/vkr/?program=p210232654&language=ru&text_available=yes'

# Информационные системы и технологии
#url = 'https://www.hse.ru/edu/vkr/?program=p135181606%3Bg134854853%3Bp148507311&language=ru&text_available=yes'

# КБ
#url = 'https://www.hse.ru/edu/vkr/?program=p135181592&language=ru&text_available=yes'

# Компьютерные системы и сети
#url = 'https://www.hse.ru/edu/vkr/?program=p141401714&language=ru&text_available=yes'

# Корп финансы
# url = 'https://www.hse.ru/edu/vkr/?program=g53353602%3Bp135181480&language=ru&text_available=yes'

# Математика
#url = 'https://www.hse.ru/edu/vkr/?program=p135181580%3Bp135181630%3Bp225211307%3Bp135181632%3Bg53353610%3Bg122468352&language=ru&text_available=yes'

# Математика и мат физика
#url = 'https://www.hse.ru/edu/vkr/?program=p141401844&language=ru&text_available=yes'

# Мат методы естествознания
#url = 'https://www.hse.ru/edu/vkr/?program=p135181586&language=ru&text_available=yes'

# Мат методы моделирования и комп технологии
#url = 'https://www.hse.ru/edu/vkr/?program=p141401851%3Bg74148971&language=ru&text_available=yes'

# Машинное обучение и анализ данных
#url = 'https://www.hse.ru/edu/vkr/?program=p309379215&language=ru&text_available=yes'

# Науки о данных
#url = 'https://www.hse.ru/edu/vkr/?program=g113512383%3Bp135181582&language=ru&text_available=yes'

# Прикладная информатика
#url = 'https://www.hse.ru/edu/vkr/?program=g133189697%3Bp148507309&language=ru&text_available=yes'

# Программная инженерия
#url = 'https://www.hse.ru/edu/vkr/?program=p135181642%3Bg122468465%3Bp135181646%3Bp135181644%3Bp581739103&language=ru&text_available=yes'

# Программная инженерия (о/з)
#url = 'https://www.hse.ru/edu/vkr/?program=p194016449%3Bp428437744&language=ru&text_available=yes'

# Системы управления и обработки информации в инженерии
#url = 'https://www.hse.ru/edu/vkr/?program=p135181360%3Bg74148969&language=ru&text_available=yes'

# Физика
#url = 'https://www.hse.ru/edu/vkr/?program=p222398942%3Bp192477926%3Bp222398940%3Bp412864136%3Bp192477918&language=ru&text_available=yes'

# Финансовый аналитик
#url = 'https://www.hse.ru/edu/vkr/?program=g53353178%3Bp135181472&language=ru&text_available=yes'

# Финансовый инжиниринг
#url = 'https://www.hse.ru/edu/vkr/?program=p141401781&language=ru&text_available=yes'

# Финансы
#url = 'https://www.hse.ru/edu/vkr/?program=p135181476%3Bp135181486%3Bg53353174%3Bg53353176%3Bp135181490%3Bg53353175%3Bp135181488%3Bg71611638%3Bp412864134&language=ru&text_available=yes'

# Экономика
url = 'https://www.hse.ru/edu/vkr/?program=g134854818%3Bp135181668%3Bp135181666%3Bp135181670%3Bg122468455%3Bp135181656%3Bg122392180%3Bp135180914%3Bp148507174%3Bg53353171%3Bg122468459%3Bg122325655%3Bg74148965%3Bp135181406%3Bp581739166&language=ru&text_available=yes'

# Экономика (в)
# url = 'https://www.hse.ru/edu/vkr/?program=p135181686&language=ru&text_available=yes'

# Экономика (з)
# url = 'https://www.hse.ru/edu/vkr/?program=p135181688&language=ru&text_available=yes'


def process_page(page_num):
    print(f'PAGE = {page_num}')
    page = requests.get(f'{url}&page={page_num}')
    soup = BeautifulSoup(page.text, "html.parser")
    links = soup.find_all('a', class_='link')
    base_url = "https://www.hse.ru"
    vkr_links = [base_url + link['href'] for link in links if 'href' in link.attrs
                 and link['href'] != '/edu/vkr/' and link['href'].startswith('/edu/vkr/')
                 and 'page' not in link['href']]

    print(f'VKR LINKS: {vkr_links}')

    for link in vkr_links:
        process_vkr_link(link)

def process_vkr_link(link):
    response = requests.get(link)
    soup1 = BeautifulSoup(response.text, "html.parser")
    topic = soup1.find('h1', class_='vkr__title vkr__title-page').get_text()
    faculty = soup1.find_all('p', class_='vkr-card__item')
    for item in faculty:
      if 'Программа:' in item.get_text():
          program_span = item.find('span', class_='vkr-card__value')
          program = program_span.get_text().strip()
          next_span = program_span.find_next_sibling('span', class_='vkr-card__value')
          if next_span:
              program += f" {next_span.get_text().strip()}"
    if soup1.find_all('p', class_='vkr-card__item 1'):
        score = int(soup1.find_all('p', class_='vkr-card__item 1')[0].text.split()[-1])
    else:
        return

    l = soup1.find('a', class_='vkr-icon__text link link_no-underline')
    if l:
      l = l.get('href')
    else:
      return

    topic_cleaned = re.sub('[^A-Za-z0-9А-Яа-я]+', '_', topic)
    with open(f'{topic_cleaned[:100]}', 'wb') as file:
      file.write(requests.get(l).content)

    try:
        document = Document(f'{topic_cleaned[:100]}')
        fullText = [para.text for para in document.paragraphs]
        text = '\n'.join(fullText)
        new_row = {'topic': topic, 'score': score, 'faculty': program, 'text': text}
        df.loc[len(df)] = new_row
        print(f'Прочитан файл DOCX для ВКР {topic}')
    except:
        print('Файл не типа docx')

    try:
        text = extract_text(f'{topic_cleaned[:100]}')
        new_row = {'topic': topic, 'score': score, 'faculty': program, 'text': text}
        df.loc[len(df)] = new_row
        print(f'Прочитан файл PDF для ВКР {topic}')
    except:
        print('Файл не типа pdf')

for page_num in range(131, 141):
    process_page(page_num)
```
